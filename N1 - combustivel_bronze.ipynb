```python
df_bronze = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "false") \
    .option("sep", ";") \
    .option("encoding", "UTF-8") \
    .load("/Volumes/playgroundlearn/combustivel_brasil/base/anp.csv")

# Renomeia colunas pra remover caracteres especiais
df_bronze = df_bronze.toDF(
    "regiao_sigla",
    "estado_sigla", 
    "municipio",
    "revenda",
    "cnpj_revenda",
    "nome_rua",
    "numero_rua",
    "complemento",
    "bairro",
    "cep",
    "produto",
    "data_coleta",
    "valor_venda",
    "valor_compra",
    "unidade_medida",
    "bandeira",
    "regiao_sigla2"
)

# Salva como Delta
df_bronze.write.format("delta") \
    .mode("overwrite") \
   .saveAsTable("playgroundlearn.combustivel_brasil.bronze_gasolina")

print("Bronze salvo com sucesso!")
```
